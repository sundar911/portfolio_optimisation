
#  Financial Analysis

## Daily Percentage Change

```python
tesla['returns'] = tesla['Close']/tesla['Close'].shift(1) - 1
tesla
```

```python
#ford['returns']=ford['Close']/ford['Close'].shift(1) - 1
ford['returns'] = ford['Close'].pct_change(1)
ford
```

```python
gm['returns']=gm['Close']/gm['Close'].shift(1) - 1
gm
```

```python
tesla['returns'].plot(kind='hist',bins=50)
```

```python
ford['returns'].plot(kind='hist',bins=50)
```

```python
gm['returns'].plot(kind='hist',bins=50)
```

```python
tesla['returns'].plot(kind='hist',bins=50,alpha=0.5,label='tesla',figsize=(10,8))
ford['returns'].plot(kind='hist',bins=50,alpha=0.5,label='ford')
gm['returns'].plot(kind='hist',bins=50,alpha=0.5,label='gm')
plt.legend()
```

```python
tesla['returns'].plot(kind='kde',label='tesla',figsize=(10,8))
ford['returns'].plot(kind='kde',label='ford')
gm['returns'].plot(kind='kde',label='gm')
plt.legend()
#ford looks the most stable 
#tesla looks the most volatile as it's the widest
```

```python
dret = pd.concat([tesla['returns'],ford['returns'],gm['returns']],axis=1)
dret.columns=['tesla','ford','gm']
dret.plot(kind='box',figsize=(8,8))
```

## Comparing Daily Returns between Stocks

```python
scatter_matrix(dret, alpha=0.2, figsize=(8,8), hist_kwds={'bins':50})
plt.tight_layout();
```

Looks like Ford and GM have some sort of possible relationship

```python
dret.plot(kind='scatter',x='gm',y='ford',figsize=(10,8),alpha=0.5)
```

____
## Cumulative Daily Returns

```python
tesla['Cumulative Return'] = (1 + tesla['returns']).cumprod()
ford['Cumulative Return'] = (1 + ford['returns']).cumprod()
gm['Cumulative Return'] = (1 + gm['returns']).cumprod()
```

```python
tesla['Cumulative Return'].plot(label='tesla',figsize=(16,6))
ford['Cumulative Return'].plot(label='ford')
gm['Cumulative Return'].plot(label='gm')

plt.title("Cumulative Return")
plt.legend();
```
